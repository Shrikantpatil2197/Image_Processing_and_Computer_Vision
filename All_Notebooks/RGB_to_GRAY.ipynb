### Q.1 write python code  from scratch to convert color image (Rose.jpg)  to grayscale image using  a def function and list of comprehension method and plot processing time comparison with other methods.[input image: Rose.jpg ]

```python
#function declaration:
def colortogray(img_bgr):
#METHOD 1
    print('\tMethod 1 results:')
    t_start1 = time.time()
    b = img_bgr[:,:,0]
    g = img_bgr[:,:,1]
    r = img_bgr[:,:,2]
    prop = img_bgr.shape
#using list comprehension:
    img_gray0 = [[(0.2989*r[i][j])+(0.5870*g[i][j])+(0.114*b[i][j])for i in range(prop[0])]for j in range(prop[1])]
    t_end1 = time.time()
    time_forloop = t_end1-t_start1
    print('Total processing time in seconds for method 1:',time_forloop)
    plt.imshow(img_gray0)
    plt.title('greyscale image (method-1)')
    plt.show()  

#METHOD 2
    print('\tMethod 2 results:')
    t_start2 = time.time()
    img_gray = cv2.cvtColor(img_bgr,cv2.COLOR_BGR2GRAY)
    t_end2 = time.time()
    time_cv2 = t_end2-t_start2
    print('Total processing time in seconds for method 2:',time_cv2)
    plt.imshow(img_gray)
    print('shape:',img_gray.shape)
    plt.title('greyscale image (method-2)')
    plt.show() 

#METHOD 3
    print('\tMethod 3 results:')
    t_start3 = time.time()
    img_gray1 = np.mean(img_bgr, axis=2)
    t_end3 = time.time()
    time_mean = t_end3-t_start3
    print('Total processing time in seconds for method 3:',time_mean)
    plt.imshow(img_gray1)
    print('shape:',img_gray1.shape)
    plt.title('greyscale image (method-3)')
    plt.show()
    
#comparision
    print('\tProcessing time comparision:\n')
    x = ['METHOD-1','METHOD-2','METHOD-3']
    proc_times = [time_forloop, time_cv2, time_mean]

    x_pos = [i for i in range(len(x))]
    plt.bar(x_pos, proc_times, color='green')
    plt.xlabel('<------------------METHOD----------------->')
    plt.ylabel('<------------------PROCESSING TIME----------------->')
    plt.title('processing time vs Method')
    plt.xticks(x_pos,x)
    plt.show()

    
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
import cv2
import time
print('Input Image:')
#image = input('Enter valid color image name:') #Rose.jpg for this assignment
#img_bgr = cv2.imread(image)
img_bgr = cv2.imread('Rose.jpg')
plt.imshow(img_bgr)
print(img_bgr.shape)
plt.title('input BGR image:\n')
plt.show()
#function call:
colortogray(img_bgr)
```
