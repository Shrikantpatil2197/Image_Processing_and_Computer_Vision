### Q3.  Write python code  from scratch to  add and subtract two images  using def function and list of comprehension method and plot processing time comparison with other methods.[input images: image1.jpg and image2.jpg]

```python
def addition(img1,img2):        #function declaration
#method-1    
    print('\t addition results:')
    t_start1 = time.time()
    addition1 = [[img1[i][j]+img2[i][j] for j in range(len(img1[0]))]for i in range(len(img1))]
    t_end1 = time.time()
    time_forloop = t_end1 - t_start1
    print('Total processing time in seconds for method 1:',time_forloop)
    plt.imshow(addition1)  
    plt.title('after addition(method-1)')
    plt.show()
    
#method-2    
    t_start2 = time.time()
    addition2 = cv2.add(img1,img2)
    t_end2 = time.time()
    time_cv2 = t_end2 - t_start2
    print('Total processing time in seconds for method 1:',time_cv2)
    plt.imshow(addition2)  
    plt.title('after addition(method-2)')
    plt.show()
    
#comparision    
    print('\tProcessing time comparision(for addition):\n')
    x = ['METHOD-1','METHOD-2']
    proc_times = [time_forloop, time_cv2]
    x_pos = [i for i in range(len(x))]
    plt.bar(x_pos, proc_times, color='green')
    plt.xlabel('<------------------METHOD----------------->')
    plt.ylabel('<------------------PROCESSING TIME----------------->')
    plt.title('processing time vs Method')
    plt.xticks(x_pos,x)
    plt.show()
    
    
def subtraction(img1,img2):           #function declaration
#method-1    
    print('\t subtraction results:')
    t_start1 = time.time()
    subtraction1 = [[img1[i][j]-img2[i][j] for j in range(len(img1[0]))]for i in range(len(img1))]
    t_end1 = time.time()
    time_forloop = t_end1 - t_start1
    print('Total processing time in seconds for method 1:',time_forloop)
    plt.imshow(subtraction1)  
    plt.title('after subtraction(method-1)')
    plt.show()
    
#method-2    
    t_start2 = time.time()
    subtraction2 = cv2.subtract(img1,img2)
    t_end2 = time.time()
    time_cv2 = t_end2 - t_start2
    print('Total processing time in seconds for method 1:',time_cv2)
    plt.imshow(subtraction2)  
    plt.title('after subtraction(method-2)')
    plt.show()
    
#comparision    
    print('\tProcessing time comparision(for subtraction):\n')
    x = ['METHOD-1','METHOD-2']
    proc_times = [time_forloop, time_cv2]
    x_pos = [i for i in range(len(x))]
    plt.bar(x_pos, proc_times, color='green')
    plt.xlabel('<------------------METHOD----------------->')
    plt.ylabel('<------------------PROCESSING TIME----------------->')
    plt.title('processing time vs Method')
    plt.xticks(x_pos,x)
    plt.show()
    
    
    
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
import cv2
import time
img1 = cv2.imread('input1.jpg')
plt.imshow(img1)
plt.title('input image-1')
plt.show()
img2 = cv2.imread('input2.jpg')
plt.imshow(img2)
plt.title('input image-2')
plt.show()
addition(img1,img2)          #function call
subtraction(img1,img2)      #function call
```
