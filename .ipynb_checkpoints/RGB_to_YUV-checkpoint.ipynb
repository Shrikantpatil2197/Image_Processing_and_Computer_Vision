```python
def colortoyuv(img):
    r = img[:,:,0]
    g = img[:,:,1]
    b = img[:,:,2]
    prop = img.shape
#using list comprehension:
    y = [[(0.2989*r[i][j])+(0.5870*g[i][j])+(0.114*b[i][j])for j in range(prop[0])]for i in range(prop[1])]
    u = [[(-0.147*r[i][j])+(-0.289*g[i][j])+(0.436*b[i][j])for j in range(prop[0])]for i in range(prop[1])]
    v = [[(0.615*r[i][j])+(-0.515*g[i][j])+(-0.100*b[i][j])for j in range(prop[0])]for i in range(prop[1])]
    img_yuv = np.dstack((y,u,v))
#     plt.imshow(img_yuv)
    plt.imshow(img_yuv.astype('uint8'))
    plt.title('yuv image')
    plt.show()  
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
import cv2
import time
print('Input Image:')
img = img.imread('lena.tif')
plt.imshow(img)
print(img.shape)
plt.title('input image:\n')
plt.show()
#function call:
colortoyuv(img)
```
