### Q 2.Write python code  from scratch to convert Grayscale  image to Binary images using a def function and list of comprehension method and  plot processing time comparison with other methods.[input image: Rose.jpg]

```python
def greytobinary(img_gray):    #function declaration  #from greyscale to binary
#METHOD-1
    print('\tMethod 1 results:')
    t_start1 = time.time()
    prop = img_gray.shape
    img_gray0 = [[0 if img_gray[i][j]<=127 else 255 for i in range(prop[0])] for j in range(prop[1])] #list comprehension
    t_end1 = time.time()
    time_forloop = t_end1-t_start1
    print('Total processing time in seconds for method 1:',time_forloop)
    plt.imshow(img_gray0)
    plt.title('Binary Image')
    plt.show()

#METHOD 2
    print('\tMethod 2 results:')
    img_gray1 = cv2.cvtColor(img_bgr,cv2.COLOR_BGR2GRAY)
    t_start2 = time.time()
    (thresh,img_blacknwhite) = cv2.threshold(img_gray1,127,255,0)
    t_end2 = time.time()
    time_cv2 = t_end2-t_start2
    print('Total processing time in seconds for method 2:',time_cv2)
    plt.imshow(img_blacknwhite)
    print('shape:',img_blacknwhite.shape)
    plt.title('greyscale image (method-2)')
    plt.show() 
    
#comparision
    print('\tProcessing time comparision:\n')
    x = ['METHOD-1','METHOD-2']
    proc_times = [time_forloop, time_cv2]

    x_pos = [i for i in range(len(x))]
    plt.bar(x_pos, proc_times, color='green')
    plt.xlabel('<------------------METHOD----------------->')
    plt.ylabel('<------------------PROCESSING TIME----------------->')
    plt.title('processing time vs Method')
    plt.xticks(x_pos,x)
    plt.show()

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
import cv2
import time
print('\tInput Image:')
#image = input('Enter valid color image name:') #Rose.jpg for this assignment
#img_bgr = cv2.imread(image)
img_bgr = cv2.imread('Rose.jpg')
plt.imshow(img_bgr)
print('shape:',img_bgr.shape)
plt.title('input BGR image:\n')
plt.show()
print('\tgreyscale image:')
img_gray = cv2.cvtColor(img_bgr,cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray)
print('shape:',img_gray.shape)
plt.title('greyscale image')
plt.show()
greytobinary(img_gray) #function call
```
