```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

#Reading the image
img = cv2.imread('input_1.png')
plt.imshow(img)
plt.show()
```

```python
#Extraction of each color chanel

r = img[:, :, 0]
g = img[:, :, 1]
b = img[:, :, 2]
#plt.imshow(b)
#plt.show()
```

```python
#Calculate the Hue

import math
import numpy as np

#eps = np.finfo(float).eps
th = np.arccos((0.5 * ((r - g) + (r - b))) / ((np.sqrt((r - g) ** 2 + (r - b) * (g - b)))+np.spacing(1)))
H = th

if (b > g).all():
    H = 2 * np.pi - H
    
H = H / (2 * np.pi)
```

```python
#Calculate the Saturation

# S = 1 - 3 * any(min(any(min(r, g)), b)) / (r + g + b+np.spacing(1))
S=1-3*(min(min(r.all(),g.all()),b.all()))/(r.all()+g.all()+b.all()+np.spacing(1))
```

```python
#Calculate the Intensity (the mean of colors)

I = (r + g + b) / 3
```

```python
#The resulting HSI ( concatenating the Hue Saturation and Intensity )

hsi = np.concatenate(3, H, S, I)
```

```python
Value = hsi[:, :, 2]
    
#Calculate the fft of the Intensity (Value)

DFT2d_Value = np.fft.fft2(Value)
```

```python
#Calculate the Gaussian

M , N = Value.shape
sigma = 4
Nr = np.fft.fftshift((-M / 2), ((M / 2) - 1))
Nc = np.fft.ifftshift((-N / 2), ((N / 2) - 1))
Nr,Nc = np.meshgrid(Nr,Nc)
dft_gauss_kernel = exp(-2 * sigma ^ 2 * np.pi ^ 2 * ((Nr / M) ^ 2 + (Nc / N ) ^ 2))
```
