```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math

def SimplestColorBalanceRGB(image, s):
    imagedata = np.float64(image)
    print(imagedata.shape)
    plt.imshow(image/255)
    plt.title('Original')
    pagesize = int(np.size(imagedata)/3)
    vector = np.reshape(imagedata,(pagesize,3))
    vector_n = np.sort(vector)
    min = vector_n[0,:]
    max = vector_n[pagesize-1,:]
    s1 = s/2
    s2 = s1
    pos_v_min = math.floor(pagesize * s1/100)
    pos_v_max = math.floor((pagesize*(1-(s2/100)))-1)
    v_min = vector_n[pos_v_min-1,:]
    v_max = vector_n[pos_v_max-1,:]
    for i in range(pagesize):
        if vector_n[i,0] > v_max[0]:
            vector_n[i,0] = v_max[0]
        elif vector_n[i,0] < v_min[0]:
            vector_n[i,0] = v_min[0]
    for i in range(pagesize):
        if vector_n[i,1] > v_max[1]:
            vector_n[i,1] = v_max[1]
        elif vector_n[i,1] < v_min[1]:
            vector_n[i,1] = v_min[1]    
    for i in range(pagesize):
        if vector_n[i,2] > v_max[2]:
            vector_n[i,2] = v_max[2]
        elif vector_n[i,2] < v_min[2]:
            vector_n[i,2] = v_min[2]  
    for i in range(3):       
        for j in range(pagesize):
            vector_n[i] = ((vector_n[i]-v_min) * (255-0) )/((v_max-v_min) + 0 )
    
    result = np.reshape(vector_n,imagedata.shape)
    plt.imshow(result/255)
    plt.title('result')
      
img_bgr = cv2.imread('input_0.png')
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
print(img_rgb.shape)
plt.imshow(img_rgb)
plt.show()
SimplestColorBalanceRGB(img_rgb,10)

```
